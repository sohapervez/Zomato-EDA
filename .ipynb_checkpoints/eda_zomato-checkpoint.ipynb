# Missing values

```python
# One way

df.isnull().sum() #how many features have missing values
```

```python
# Another way to find missing values
# This will directly display the features that have null values greater than 0

[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
# Visualizing the null values

sns.heatmap(df.isnull(), yticklabels = False, cbar = False, cmap = 'viridis')
```

In the above heatmap, we can't see the null values (which are 9 for cuisines) because we have 9551 rows
and among them 9 won't be much visible. If it was a bigger number then we would have seen it.

```python
# checking the other dataset

df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
# In both df and df_country we have one feature common i.e. Country Code
# Combining both dataframes on 'Country Code' column

final_df = pd.merge(df, df_country, on = 'Country Code', how = 'left')
```

```python
final_df.head(2)
```

```python
# checking data types

final_df.dtypes
```

```python
final_df.columns
```

```python
# finding out number of records w.r.t. each country

final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
final_df.Country.value_counts().index #get all the country names
```

```python
country_val = final_df.Country.value_counts().values # get values for those countries
```

```python
# Pie Chart - Top 3 countries that uses zomato

plt.pie(country_val[:3], labels = country_names[:3], autopct = '%1.2f%%')
```

Observation: Zomato's maximum records/transactions are from India, then the USA and the UK.

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0: 'Rating Count'})
```

```python
ratings.head()
```

# Observation

1. Rating 4.5 - 4.9 ---> excellent.
2. Rating 4.0 - 4.4 ---> very good.
3. Rating 3.5 - 3.9 ---> good.
4. Rating 2.5 - 3.4 ---> average.
5. Rating 1.8 - 2.4 ---> poor.
6. 2148 people haven't rated.

```python
# Visualizing the above observation

import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6) # increasing the plot size
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', data = ratings)
```

```python
# Incorporating 'Rating color' in the plot so that visually it's easy to understand which color corresponds to
# which rating

sns.barplot(x = 'Aggregate rating', y = 'Rating Count', hue = 'Rating color', data = ratings, palette = ['white', 'red', 
                                                                                                         'orange', 'yellow', 'green', 'green'])
```

# Observation

1. Not rated count is very high
2. The maximum number of ratings are between 2.5 and 3.4

```python
# Count plot
# Plotting w.r.t. categorical variables

sns.countplot(x = 'Rating color', data = ratings, palette = ['white', 'red', 'orange', 'yellow', 'green', 'green'])
```

```python
# Creating a pie chart w.r.t. top 5 cities

city_names = final_df.City.value_counts().index
city_val = final_df.City.value_counts().values # get values for those countries

plt.pie(city_val[:5], labels = city_names[:5], autopct = '%1.2f%%')
```

```python
# Finding the top 5 cuisines

# Split the Cuisines column to handle multiple cuisines per restaurant
cuisine_series = final_df['Cuisines'].str.split(', ').explode()

# Count the occurrences of each cuisine
cuisine_counts = cuisine_series.value_counts()

# Get the top 5 cuisines
top_5_cuisines = cuisine_counts.head(5)

print(top_5_cuisines)
```

# Observation

The most famous cuisine is North Indian.
